```python
index_ct_death=np.append(index_ct,15)
index_ct_death
```

```python
df_ct_death=df_data.iloc[:,index_ct_death]
df_ct_death
```

# CT + CD + Death

```python
index_ct_cd_death=np.append(np.append(index_ct,index_cd),15)
df=df_data.iloc[:,index_ct_cd_death]
```

```python
df=df.drop(['Alcohol abuse'],axis=1)
df
```

```python
df['DEATH [d from CT]']=df['DEATH [d from CT]'].fillna(0)
df
```

```python
df=df.dropna()
df
```

```python
df_one = pd.get_dummies(df['Sex'])
 
# display result
df_two = pd.concat((df_one, df), axis=1)
df_two = df_two.drop(["Sex"], axis=1)
df_two = df_two.drop(["Male"], axis=1)
df = df_two.rename(columns={"Female": "Gender"})
```

```python
df_one = pd.get_dummies(df['BMI >30'])
 
# display result
df_two = pd.concat((df_one, df), axis=1)
df_two = df_two.drop(['BMI >30'], axis=1)
df_two = df_two.drop(["N"], axis=1)
df = df_two.rename(columns={"Y": 'BMI >30'})
```

```python
df_one = pd.get_dummies(df['Tobacco'])
 
# display result
df_two = pd.concat((df_one, df), axis=1)
df_two = df_two.drop(['Tobacco'], axis=1)
df_two = df_two.drop(["No"], axis=1)
df = df_two.rename(columns={"Yes": 'Tobacco'})
```

```python
df_one = pd.get_dummies(df['Met Sx'])
 
# display result
df_two = pd.concat((df_one, df), axis=1)
df_two = df_two.drop(['Met Sx'], axis=1)
df_two = df_two.drop(["N"], axis=1)
df = df_two.rename(columns={"Y": 'Met Sx'})
```

```python
df=df.drop(['FRS 10-year risk (%)'],axis=1)
df
```

```python
df=df.apply(lambda s: pd.to_numeric(s, errors='coerce')).dropna()
df
```

```python
df_num=(df-df.min())/(df.max()-df.min())
df_num
```

```python
data_ctcd_death_np=df_num.values
```

```python
np.save('..\Data\data_ctcd_death_np.npy',data_ctcd_death_np)
```
